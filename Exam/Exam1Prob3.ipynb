```python
from __future__ import division
from statistics import mode
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn import tree
import numpy as np
import pandas as pd
import os
from PIL import Image
import graphviz
from sklearn.externals.six import StringIO
import pydotplus

path = os.getcwd()
datasetpath = os.path.join(path,"data")
datasetpath = os.path.join(datasetpath,"prob3.csv")


prob3df = pd.read_csv(datasetpath)
pro3df_pos = prob3df.drop(["-",],axis=1)
pro3df_neg = prob3df.drop(["+",],axis=1)





train_pos = prob3df.drop(["+","-"],axis=1)
train_neg = prob3df.drop(["-","+"],axis=1)

labels_neg = prob3df["-"]
labels_pos = prob3df["+"]
```

```python
dt_pos = tree.DecisionTreeClassifier()
# training
dt_pos.fit(train_pos,labels_pos)

dot_data_pos = StringIO()
dot_data_neg = StringIO()
# importing trees to graphviz dot format
tree.export_graphviz(dt_pos,out_file=dot_data_pos,
                     feature_names=["x","y"],
                     class_names=["0","1"],
                     filled=True,
                     rounded=True,impurity=True)




dt_neg = tree.DecisionTreeClassifier()
# training
dt_neg.fit(train_neg,labels_neg)

dot_data_neg = StringIO()
# importing trees to graphviz dot format
tree.export_graphviz(dt_neg,out_file=dot_data_neg,
                     feature_names=["x","y"],
                     class_names=["0","1"],
                     filled=True,
                     rounded=True,impurity=True)

```

```python

decision_tree_graph= pydotplus.pydotplus.graph_from_dot_data(dot_data_pos.getvalue())
decision_tree_graph.write_png("dt_pos.png")
img =mpimg.imread("/Users/francisco/Desktop/Fall_2018/Data_Mining/Exam/dt_pos.png")
plt.imshow(img,interpolation='nearest',aspect="auto")
```
