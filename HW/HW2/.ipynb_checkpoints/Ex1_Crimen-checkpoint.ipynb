```python
# imports and reading data
import numpy as np
import pandas as pd
import math as m
import os
from sklearn.feature_selection import SelectKBest, chi2, SelectPercentile, f_classif, mutual_info_classif



#path
path = os.getcwd()
path = os.path.join(path,"datasets")
path = os.path.join(path,"crimen.txt")




# variable of interest (Crime Rate)
var_resp = 'R'


#pandas data frame
data = pd.read_csv(path,sep="\t")

#getting crime rate column
data_var = pd.DataFrame(data[var_resp])


data
```

```python
# Correlation between variables and the variable of interest
# Pearson correlation coefficient matrix

      
def get_corr(resp, data):
    cor_vec = []
    col_vec = []
    for col in data.drop(resp,axis=1):
        cor = np.corrcoef(data[resp],data[col])[0][1]
        print("Pearson correlation between " + resp +" and " + col+" is: ",cor )
        cor_vec.append(cor)
        col_vec.append(col)
    return cor_vec,col_vec
cor = get_corr('R',data)
print('\n')
print("The max correlation is between R and Ex0: ",np.max(cor[0]))
```
